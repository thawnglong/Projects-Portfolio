```python
nokianull = isnull[(df['Tên sản phẩm'].str.contains('nokia'))]
```

```python
# fillna nokia
nokianull['Thương hiệu'] = nokianull['Thương hiệu'].fillna('nokia')
```

```python
poconull = isnull[(df['Tên sản phẩm'].str.contains('POCO'))]
```

```python
# fillna poco
poconull['Thương hiệu'] = poconull['Thương hiệu'].fillna('poco')
```

```python
newdf = notnull.append(samsungnull)
```

```python
newdf = newdf.append(xiaominull)
```

```python
newdf = newdf.append(nokianull)
```

```python
newdf.info()
```

```python
def country_define(x):
  if x in ['lg','samsung']:
    return 'Hàn Quốc'
  elif x == 'vsmart':
    return 'Việt Nam'
  elif x == 'asus':
    return 'Đài Loan'
  elif x == 'infinix':
    return 'Hongkong'
  elif x in ['google','apple']:
    return 'Mỹ'
  elif x in['sony','vertu','nokia']:
    return 'Nhật Bản'
  else:
    return 'Trung Quốc'
```

```python
newdf['Quốc Gia'] = newdf['Thương hiệu'].apply(country_define)
```

```python
def platform_define(x):
  if x[0] == '1':
    return 'Shopee'
  elif x[0] == '2':
    return 'Lazada'
  else:
    return 'Tiki'
```

```python
newdf['Sàn'] = newdf['Mã sản phẩm'].apply(platform_define)
```

```python
newdf = newdf.drop(columns='Mã sản phẩm')
```

## Câu 4

```python
Samsung_Shopee = newdf[(newdf['Sàn']=='Shopee')&(newdf['Thương hiệu']=='samsung')]
```

```python
Samsung_Shopee
```

```python
def s23_define(x):
  if 'S23' in x :
    return 'S23'
  else:
    return 'Khác'
```

```python
Samsung_Shopee['Phân loại'] = Samsung_Shopee['Tên sản phẩm'].apply(s23_define)
```

```python
cau4 = Samsung_Shopee.groupby('Phân loại')['Tổng doanh số','Tổng số đã bán'].sum()
```

```python
cau4['% đã bán'] = cau4['Tổng số đã bán']/cau4['Tổng số đã bán'].sum()*100
```

```python
cau4['% doanh số'] = cau4['Tổng doanh số']/cau4['Tổng doanh số'].sum()*100
```
